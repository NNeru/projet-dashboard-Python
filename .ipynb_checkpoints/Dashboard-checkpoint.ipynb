```python
catnat2016=catnat[catnat['Année']=='2016']
catnat2017=catnat[catnat['Année']=='2017']
catnat2018=catnat[catnat['Année']=='2018']
catnat2019=catnat[catnat['Année']=='2019']
catnat2020=catnat[catnat['Année']=='2020']
```

```python
nb_catnat2016=catnat2016.groupby(['Région','Type_de_catnat'])[['Num_risque']].count().reset_index()
nb_catnat2016=nb_catnat2016[28:]
nb_catnat2016=nb_catnat2016.rename(columns={'Num_risque':'nb_catnat'})
nb_catnat2016=nb_catnat2016.groupby('Région')[['nb_catnat']].sum().reset_index()
```

```python
nb_catnat2017=catnat2017.groupby(['Région','Type_de_catnat'])[['Num_risque']].count().reset_index()
nb_catnat2017=nb_catnat2017[9:]
nb_catnat2017=nb_catnat2017.rename(columns={'Num_risque':'nb_catnat'})
nb_catnat2017=nb_catnat2017.groupby('Région')[['nb_catnat']].sum().reset_index()
```

```python
nb_catnat2018=catnat2018.groupby(['Région','Type_de_catnat'])[['Num_risque']].count().reset_index()
nb_catnat2018=nb_catnat2018[213:]
nb_catnat2018=nb_catnat2018.rename(columns={'Num_risque':'nb_catnat'})
nb_catnat2018=nb_catnat2018.groupby('Région')[['nb_catnat']].sum().reset_index()
```

```python
nb_catnat2019=catnat2019.groupby(['Région','Type_de_catnat'])[['Num_risque']].count().reset_index()
nb_catnat2019=nb_catnat2019[30:]
nb_catnat2019=nb_catnat2019.rename(columns={'Num_risque':'nb_catnat'})
nb_catnat2019=nb_catnat2019.groupby('Région')[['nb_catnat']].sum().reset_index()
nb_catnat2019.head()
```

```python
nb_catnat2020=catnat2020.groupby(['Région','Type_de_catnat'])[['Num_risque']].count().reset_index()
nb_catnat2020=nb_catnat2020[47:]
nb_catnat2020=nb_catnat2020.rename(columns={'Num_risque':'nb_catnat'})
nb_catnat2020=nb_catnat2020.groupby('Région')[['nb_catnat']].sum().reset_index()
nb_catnat2020.head()
```

```python
app = dash.Dash(__name__)
app.layout = html.Div([
    dcc.Dropdown(
        id='nb_catnat_year_dropdown',
        options=[
            {'label': '2016', 'value': 2016},
            {'label': '2017', 'value': 2017},
            {'label': '2018', 'value': 2018},
            {'label': '2019', 'value': 2019},
            {'label': '2020', 'value': 2020}
        ],
        value=2016
    ),
    html.Div(id='dd-output-container')
])


@app.callback(
    dash.dependencies.Output('nb_catnat_year_dropdown', 'children'),
    [dash.dependencies.Input('demo-dropdown', 'value')])
def update_output(value):
    return 'You have selected "{}"'.format(value)


if __name__ == '__main__':
    app.run_server(debug=True)
```
